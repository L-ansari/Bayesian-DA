# Bayesian Data Analysis: Distributions and Priors

This notebook explores probability distributions and prior specification for Bayesian analysis. We'll use the PreliZ library to visualize and interact with various distributions commonly used in Bayesian modeling.

```python
import preliz as pz
import numpy as np
import matplotlib.pyplot as plt
```

## Import Libraries

We'll use:
- **PreliZ**: A Python library for prior elicitation and probability distribution visualization

```python
pz.BetaBinomial(alpha=10, beta=10, n=6).plot_interactive()
```

## Beta-Binomial Distribution

The Beta-Binomial distribution is a compound distribution useful for modeling overdispersed count data. It combines:
- A **Beta distribution** for the probability parameter
- A **Binomial distribution** for the count outcome

Here we create an interactive plot with:
- `alpha=10, beta=10`: Creates a symmetric Beta prior centered at 0.5
- `n=6`: Number of trials in the binomial process



```python
pz.BetaBinomial(alpha=10, beta=10, n=6).rvs()

```

## Generating Random Samples

Let's generates a single random sample from the Beta-Binomial distribution. The `rvs()` method (random variates sample) draws random values from the distribution, which is useful for simulation and posterior predictive checks.

```python
plt.hist(pz.BetaBinomial(alpha=2, beta=5, n=5).rvs(1000)) 
pz.BetaBinomial(alpha=2, beta=5, n=5).plot_pdf()

```

## Visualizing Distribution with Different Parameters

This cell demonstrates two complementary visualizations:
1. **Histogram of 1000 samples** - Shows the empirical distribution from random samples with `alpha=2, beta=5, n=5`
2. **Probability Density Function (PDF)** - Overlays the theoretical distribution

The asymmetric parameters (alpha=2, beta=5) create a left-skewed distribution, useful for modeling scenarios where lower count values are more probable.
